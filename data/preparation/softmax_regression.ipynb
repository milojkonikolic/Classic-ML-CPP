```python
import os
import pandas as pd
```

```python
data_path = "../datasets/weather_classification_data.csv"
data = pd.read_csv(data_path)
```

```python
data.head()
```

```python
cat_features = ["Cloud Cover", "Season", "Location"]
```

```python
for cat_feature in cat_features:
    print(f"{cat_feature}: {data[cat_feature].unique()}")
```

```python
data["Weather Type"].unique()
```

```python
data_encoded = pd.get_dummies(data, columns=["Cloud Cover", "Season", "Location"])
```

```python
data_encoded["Weather Type"], weather_classes = pd.factorize(data["Weather Type"])
```

```python
# Move 'Weather Type' column to the end of the DataFrame
data_encoded = data_encoded[[col for col in data_encoded.columns if col != "Weather Type"] + ["Weather Type"]]
```

```python
data_encoded = data_encoded.astype(float)
```

```python
data_encoded.head()
```

```python
data_encoded.to_csv("../datasets/weather_classification_data_prepared.csv", index=False, header=False)
```
